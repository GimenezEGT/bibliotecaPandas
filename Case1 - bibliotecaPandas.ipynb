```python
import pandas as pd
import seaborn as sns
```

```python
varejo = pd.read_excel("varejo.xlsx")
varejo.head()
```

```python
varejo.info()
```

```python
varejo.groupby(by="idcanalvenda").idcompra.nunique()
```

```python
varejo['idcanalvenda'] = varejo['idcanalvenda'].replace("APP", "Aplicativo")
varejo.groupby(by="idcanalvenda").idcompra.nunique()
```

```python
varejo.groupby(by="bandeira").idcompra.nunique()
```

```python
varejo.groupby(by="Nome_Departamento").idcompra.nunique()
```

```python
varejo['Nome_Departamento'] = varejo["Nome_Departamento"].apply(lambda x: x.replace(" ", "_"))
```

```python
varejo.groupby("Nome_Departamento").idcompra.nunique()
```

```python
varejo[varejo["estado"].isnull()]
```

```python
vendas_semnulos = varejo.dropna()
vendas_semnulos.info()
```

```python
varejo["estado"].fillna("MS", inplace=True)
varejo.info()
```

```python
varejo["Preço"].fillna(varejo["Preço"].mean(), inplace=True)
```

```python
media_preco = varejo["Preço"].mean()
varejo["Preço"]
```

```python
vendas_correto = varejo.query("Preço < Preço_com_frete")
vendas_correto
```

```python
vendas_correto["mês"] = vendas_correto["Data"].dt.month
vendas_correto
```

```python
vendas_correto.groupby("mês").idcompra.nunique().sort_values(ascending=False).reset_index(name="Qtd._de_vendas")
```

```python
cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
vendas_correto
```

```python
cliente.describe()
```

```python
cliente.info()
```

```python
#transformando renda em float
cliente = cliente.astype({"renda":"float"})
cliente.info
```

```python
vendas_cliente = vendas_correto.merge(cliente, how='left', on="cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_venda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_venda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(), 2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,8));
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color ='green');
plt.title("Idade média por Bandeira", loc = 'center', fontsize=18, fontweight='bold', color='black');
plt.ylabel('Média de idade');
```

```python
fig = plt.figure(figsize=(10,8));
plt.bar(agg_idcanal_venda['idcanalvenda'], agg_idcanal_venda['renda'], color='green');
plt.ylabel("Renda média");
plt.title("Média de Renda por Canal de Venda", loc='center', fontsize=18,  fontweight='bold', color='black');
```

```python
venda_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(10,8))
plt.xlabel("Data da venda")
plt.ylabel("Quantidade de vendas")
x = venda_por_data['Data']
y = venda_por_data['idcompra']

plt.plot(x, y, color='green', linewidth='2');
```

```python
agg_dpto_preco = round(vendas_correto.groupby(['Nome_Departamento'])['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
fig = plt.figure(figsize=(40,20));
plt.bar(agg_dpto_preco['Nome_Departamento'], agg_dpto_preco['Preço_com_frete'], color='green');
plt.ylabel("Preço com frete");
plt.title("Média de Preço por Departamento", loc='center', fontsize=18,  fontweight='bold', color='black');
```
